```python
import plotly.express as px
import numpy
import pandas as pd


df = pd.read_csv("fifa_eda.csv")
```

```python
df.sample(5)
```

```python
#Elemento 1: Obtener matriz de correlación y dejarla como heat map
df_numerico = df.select_dtypes(include=['number'])
fig = px.imshow(df_numerico.corr(), text_auto= True, title= "Correlación general")
fig.show()
```

```python
#Elemento 2: Gráfico de Correlación que responda – “Cuál es la relación entre la edad y el overall?” – explicar la salida 

df_corr2 = df[["Age", "Overall"]]
fig = px.imshow(df_corr2.corr(), text_auto= True, title= "Correlación entre edad y overall")
fig.show()

#Observamos que ambas variables tienen una correlacion debil, esto nos indica que la edad no es un factor tan determinante para saber si un jugador es bueno o malo. 
#Hay veteranos con un alto Overall, jovenes con bajo Overall y viceversa. 
```

```python
#Elemento 3: Generar un gráfico de barras por club que indique el número de jugadores 

df_player = df.groupby("Club").size().rename("Players").reset_index().sort_values(by = "Players")
figbarra = px.bar(df_player, x = "Club" , y= "Players", title = "Jugadores por cada Club")
figbarra.show()
```

```python
#Elemento 4: Hacer un gráfico multipanel que indique la relación altura (height) vs skill moves, siendo la variable del panel si es zurdo o derecho 
fig = px.strip(df, x = "Height", y = "Skill Moves", facet_col= "Preferred Foot", color = "Age")
fig.show()
```

```python
#Elemento 5: Un gráfico que crea importante para mostrar a los potenciales cracks
df_player2 = df
index_80 = df_player2[(df_player2["Potential"]< 85)].index
df_player2.drop(index_80, inplace= True)
px.density_heatmap(df_player2, x = "Club", y= "Potential")

#Con este grafico podemos observar que la mayoria de jugadores con Potencial mayor a 85 se encuentran en los equipos mas grandes de Europa.
#Por lo que podemos decir que esos equipos cuentan con las mejores canteras del mundo
```
